```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
heights = pd.read_csv("president_heights.csv")
heights.head()
```

```python
heights.shape
```

```python
height = np.array(heights["height(cm)"])
print(height)
```

### Summary statistics

```python
print("Mean of heights =", height.mean())
print("Standard Deviation of height =", height.std())
print("Minimum height =", height.min())
print("Maximum height =", height.max())
```

Note that in each case, the aggregation operation reduced the entire array to a single summarizing value, which gives us information about the distribution of values. We may also wish to compute quantiles:

```python
print("25th percentile =", np.percentile(height, 25))
print("Median =", np.median(height))
print("50th percentile =", np.percentile(height, 50))
print("75th percentile =", np.percentile(height, 75))
```

We see that the median height of US presidents is `182 cm`, or just shy of six feet. Of course, sometimes itâ€™s more useful to see a visual representation of this data, which we can accomplish using tools in Matplotlib

```python
import seaborn as sns
sns.set() # For using the styles

plt.hist(height)
plt.title("HeightÂ DistributionÂ ofÂ PresidentsÂ ofÂ USA")
plt.xlabel("height(cm)")
plt.ylabel("Number")
plt.show()
```

```python
# in Xkcd format (webcomic) of US presidential heights
plt.xkcd()
plt.hist(height)
plt.title("Height dist of US presidents")
plt.xlabel('height(cm)')
plt.ylabel('number');
```

###  Histogram to measure the height of most US presidents

```python
plt.figure(figsize=(12,8))
sns.distplot( a=height, hist=True, kde=False, rug=True)
plt.title("Height of US Presidents")
plt.xlabel("height(cm)")
plt.ylabel("Number of Presidents")
plt.show()

# A rug plot is a plot of data for a single quantitative variable, displayed as marks along an axis. 
# It is used to visualise the distribution of the data. 
# As such it is analogous to a histogram with zero-width bins, or a one-dimensional scatter plot.
```

Majority of US Presidents had a height of 185cm.

### Using Trendline to measure the increase/decreas in the average height of the US President

```python
# Draw regression plot - Visulalize linear regression model

plt.figure(figsize=(12,8))
sns.regplot(x=heights["order"], y=heights["height(cm)"], marker="o", 
            scatter_kws={"color":"blue","alpha":0.5,"s":50}, 
            fit_reg = True, scatter = True, ci = None) # ci: Confidence Interval
plt.title("Linear Model fit line")
plt.show()
```

Correlation is positive. Over time, there has been an increase in the average height of the US Presidents.

### Using Scatterplot to plot height of Presidents in the order of their presidential term

```python
plt.figure(figsize=(12,8))
plt.plot('order', 'height(cm)', data=heights, linestyle='-', marker='o')
plt.title("Height of US Presidents")
plt.xlabel("Presidents order")
plt.ylabel("height(cm)")
plt.show()
```

### Confidence Interval

In statistics, a **confidence interval (CI)** is a type of **estimate** computed from the observed data. This gives a range of values for an unknown **parameter** (for example, a population mean). The interval has an associated confidence level chosen by the investigator. For a given estimation in a given sample, using a higher confidence level generates a wider (i.e., less precise) confidence interval. In general terms, a confidence interval for an unknown parameter is based on sampling the **distribution** of a corresponding **estimator**.
